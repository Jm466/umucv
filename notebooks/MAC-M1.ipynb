# Instalación del entorno miniforge en Mac M1
En el caso de la nueva versión M1 de Mac, vamos a usar una instalación de Miniforge, que es una alternativa a miniconda con énfasis en la compatibilidad con la nueva arquitectura de apple.

Lo descargamos del siguiente enlace:

https://github.com/conda-forge/miniforge

En nuestro caso descargamos la versión *Miniforge3-MacOSX-arm64*, y una vez descargada la instalamos con

```
bash Miniforge3-MacOSX-arm64
``` 

y seguimos las instrucciones. Una vez instalado, creamos un entorno de conda

```
conda create --name via python==3.8.5
conda activate via
```

Y podemos instalar las librerías necesarias:

```
pip install pyqtgraph pyopengl opencv-python
pip install jupyter matplotlib scikit-learn scikit-image pydot
```

Así como el paquete con funciones de utilidad de la asignatura:
```
pip install http://robot.inf.um.es/material/umucv.tar.gz  
```

## Instalar PyQT5

Para poder instalar PyQT5 es necesario descargarse el gestor de paquetes *Homebrew*, si no lo tienes se descarga con

```
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
```

Entonces descargamos PyQT5 con

```
brew install pyqt@5
```

# Instalación de Tensorflow en Mac M1

Una vez tenemos preparado nuestro entorno, lo primero es descargar las dependencias de Tensorflow:

```
conda install -c apple tensorflow-deps
```

Y seguidamente descargamos la versión base de Tensorflow:

```
python -m pip install tensorflow-macos
```

Por último descargamos el plugin metal para Tensorflow, que acelera el funcionamiento de Tensorflow en este dispositivo, más info en: https://blog.tensorflow.org/2021/06/pluggabledevice-device-plugins-for-TensorFlow.html

```
python -m pip install tensorflow-metal
```
